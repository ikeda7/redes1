```python
import ipaddress
```

```python
def calcular_subredes(endereco_ip, num_subredes):
    endereco = ipaddress.IPv4Network(endereco_ip)
    endereco_rede = endereco.network_address
    prefixo_original = endereco.prefixlen
    
    # Calcular a potência de 2 maior ou igual ao número de sub-redes
    potencia = 1
    while 2 ** potencia < num_subredes:
        potencia += 1
    
    # Verificar se o prefixo original + potência é menor ou igual a 32
    if prefixo_original + potencia > 32:
        print("Não é possível criar esse número de sub-redes com o prefixo atual.")
        return
    
    # Calcular a máscara de sub-rede
    bits_subredes = potencia
    mascara = ipaddress.IPv4Network(f"{endereco.network_address}/{prefixo_original + bits_subredes}").netmask
    
    # Calcular e imprimir as sub-redes, ID de rede, endereço de rede e endereço de broadcast
    print(f"Endereço IP: {endereco}")
    print(f"Máscara de Sub-rede: {mascara}")
    print("Sub-redes:")
    subredes = list(endereco.subnets(new_prefix=prefixo_original + bits_subredes))
    for i, subrede in enumerate(subredes):
        id_rede = subrede.network_address
        endereco_rede = subrede.network_address + 1
        endereco_broadcast = subrede.broadcast_address
        print(f"Sub-rede {i+1}:")
        print(f"  ID de Rede: {id_rede}")
        print(f"  Endereço de Rede: {endereco_rede}")
        print(f"  Endereço de Broadcast: {endereco_broadcast}")
        print(f"  Máscara de Sub-rede: {mascara}")
        print()
```

```python
# Exemplo de uso
endereco_ip = "192.168.0.0/24"
num_subredes = 40
calcular_subredes(endereco_ip, num_subredes)
```
