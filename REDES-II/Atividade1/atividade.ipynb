```python
class IPv6Router:
    def __init__(self):
        self.routing_table = {}

    def add_route(self, country, block):
        self.routing_table[country] = block

    def route_rightmost(self, packet):
        destination = packet['destination']
        for country, block in self.routing_table.items():
            if destination.startswith(block):
                return f"Sending packet to {destination} via {country} (Rightmost)"
        return "No matching route found"

    def route_leftmost(self, packet):
        destination = packet['destination']
        for country, block in reversed(self.routing_table.items()):
            if destination.startswith(block):
                return f"Sending packet to {destination} via {country} (Leftmost)"
        return "No matching route found"

# Exemplo de uso
router = IPv6Router()
router.add_route('USA', '2001:0500::')
router.add_route('Brazil', '2804:0c00::')
router.add_route('China', '2404:0000::')
router.add_route('Germany', '2a0d::')
router.add_route('Australia', '2406:da00::')
router.add_route('Japan', '2401:0000::')

packet = {'source': '2001:0500::1234', 'destination': '2804:0c00::5678'}

print(router.route_rightmost(packet))
print(router.route_leftmost(packet))
```
